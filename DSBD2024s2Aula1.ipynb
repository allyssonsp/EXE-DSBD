# 0.5 Estatística extremamente básica

* MÉDIA  
    Soma dos valores das amostras dividida pelo total de amostras

* MEDIANA  
    Valor do "meio" do conjunto de dados numéricos, ou média dos dois valores do meio.

* MODA  
    É o primeiro valor que mais se repete no conjunto (mais frequente)

* VARIÂNCIA  
    Mede a dispersão dos dados (se espalhados ou agrupados ao redor da média); calculada pela soma dos quadrados da subtração do valor de cada amostra por sua média, dividido pelo número de amostras

* DESVIO PADRÃO  
    Raiz quadrada da *variância* da amostra

O módulo *statistics* possui essas implementações:

```python
import statistics

lista_de_valores = [0, 1, 2, 2, 3, 4, 5, 6, 7]

# Média
statistics.mean(lista_de_valores)
```

## Funções com argumentos

A entrada de uma função pode ser 0 ou mais argumentos, com alguns deles <b>mandatórios</b>, outros *default* e/ou *opcionais*

```python
def pergunta(confirma, tentativas = 3, lembrete = "Tente novamente!"):
    while True:
        ok = input(confirma)
        if ok in ('s', 'S'):
            return True
        if ok in ('n', 'N'):
            return False
        tentativas = tentativas - 1
        if tentativas < 0:
            raise ValueError("Resposta inválida!")
        print(lembrete)
```

```python
# Chamada com argumento mandatório (argumento "confirma") e demais argumentos "default"
pergunta("Deseja mesmo apagar?")
```

```python
# Chamada com argumento mandatório (argumento "confirma") e um argumento OPCIONAL "tentativas"
pergunta("Deseja mesmo apagar?", 1)
```

```python
# Chamada com todos os argumentos
pergunta("Deseja mesmo apagar?", 1, "Sim [s] ou Não [n]")
```
